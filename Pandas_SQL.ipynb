```python
# Filtering Rows
# using relational (<,<=,>,>=,!=,==) operators and boolean masking
# filtering rows in  data were rating is greater than 8.5

movies[movies['IMDB_Rating'] > 8.5]
```

###**SORTING DATA**

```python
# Sorting data (ASC OR DESC)

movies.sort_values(by = 'Genre',ascending=True) # in ASC order
```

```python
movies.sort_values(by = 'Genre',ascending=False) # in DESC order
```

###**GROUPING AND AGGREGATING**

```python
# Grouping and Aggregating

# groupby() groups all the categories into a single group and along with any Aggregation
# which actor have the highest gross?

movies.groupby('Star1')['Gross'].sum().sort_values(ascending=False)
```

```python
# groupby on the basis of two columns
# find the top 5 combo of actor and director who have the highest gross

movies.groupby(['Star1','Director'])['Gross'].sum().sort_values(ascending=False).reset_index().head(5)
```

###**JOINING/MERGE**
- **inner**
- **right**
- **left**
- **outer**

```python
# Joining DataFrames (inner_join,right_join,left_join,outer_join)
# creating our own data to perform joins
df1 = pd.DataFrame({
    'ID': [1,2,3,4],
    'Name': ['Bob','Eve','Alice','jhon']
})

df2 = pd.DataFrame({
    'ID' : [3,4,5,6],
    'Salary' : [25000,45000,60000,80000]
})

df1
```

```python
df2
```

```python
df1.merge(df2,how= 'inner',on = 'ID')
```

```python
# another syntax for joining by using pd.merge()

pd.merge(df1,df2,how= 'inner',on= 'ID')
```

```python
pd.merge(df1,df2,how= 'right',on= 'ID')
```

```python
pd.merge(df1,df2,how= 'left',on= 'ID')
```

```python
pd.merge(df1,df2,how= 'outer',on= 'ID')
```

###**CONCATING**

```python
# concating DataFrame

pd.concat([df1,df2],ignore_index=True)
```

```python
df1.append(df2,ignore_index=True)
```

###**PIVOT**

```python
# pivoting data
# creating data
data = {
  'Date': ['2023-01-01','2023-01-01','2023-01-02','2023-01-02','2023-01-01','2023-01-02'],
  'Product':['Apple','Banana','Apple','Banana','Apple','Banana'],
  'Sales': [100,150,200,120,140,80],
  'Region' : ['North','south','north','south','north','south']
}

df = pd.DataFrame(data)
```

```python
df
```

```python
# index,columns,values,aggfunc are the important field requirement
pivot_result = pd.pivot_table(df,index = ['Date','Region'],columns = 'Product',values = 'Sales',aggfunc = 'mean',fill_value = 0)
```

```python
pivot_result
```

###**APPLYING FUNCTION**

```python
# applying function
# apply()
# we will calculate the normalized gross for each movie

def calculate_normalized_gross(row):
  return round(row['Gross'] / row['Runtime'],2)
```

```python
movies['normalized_gross'] = movies.apply(calculate_normalized_gross, axis = 1)

movies
```
